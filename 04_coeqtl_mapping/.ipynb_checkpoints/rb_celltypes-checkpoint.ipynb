```python
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline

from pathlib import Path
workdir = Path("./coeqtl_mapping/")

celltypes = ['CD4T', 'CD8T', 'monocyte', 'DC', 'NK', 'B']
import matplotlib
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):
    """
    Create a heatmap from a numpy array and two lists of labels.

    Parameters
    ----------
    data
        A 2D numpy array of shape (M, N).
    row_labels
        A list or array of length M with the labels for the rows.
    col_labels
        A list or array of length N with the labels for the columns.
    ax
        A `matplotlib.axes.Axes` instance to which the heatmap is plotted.  If
        not provided, use current axes or create a new one.  Optional.
    cbar_kw
        A dictionary with arguments to `matplotlib.Figure.colorbar`.  Optional.
    cbarlabel
        The label for the colorbar.  Optional.
    **kwargs
        All other arguments are forwarded to `imshow`.
    """

    if not ax:
        ax = plt.gca()

    # Plot the heatmap
    im = ax.pcolormesh(data, **kwargs)

    # Create colorbar
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom")

    # Let the horizontal axes labeling appear on top.
    ax.tick_params(top=True, bottom=False,
                   labeltop=True, labelbottom=False)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=-30, ha="right",
             rotation_mode="anchor")

    # Turn spines off and create white grid.
#     ax.spines[:].set_visible(False)

#     ax.set_xticks(np.arange(-0.5, data.shape[1]-2, 1), minor=True)
#     ax.set_yticks(np.arange(-0.5, data.shape[0]-2, 1), minor=True)
    # Show all ticks and label them with the respective list entries.
    ax.set_xticklabels([""]+col_labels)
    ax.set_yticklabels([""]+row_labels)
#     ax.grid(which='minor', color="white", linestyle='-', linewidth=2)
#     ax.tick_params(which="minor", bottom=False, left=False)
    return im, cbar


def annotate_heatmap(im, data=None, valfmt="{x:.2f}",
                     textcolors=("black", "white"),
                     threshold=None, **textkw):
    """
    A function to annotate a heatmap.

    Parameters
    ----------
    im
        The AxesImage to be labeled.
    data
        Data used to annotate.  If None, the image's data is used.  Optional.
    valfmt
        The format of the annotations inside the heatmap.  This should either
        use the string format method, e.g. "$ {x:.2f}", or be a
        `matplotlib.ticker.Formatter`.  Optional.
    textcolors
        A pair of colors.  The first is used for values below a threshold,
        the second for those above.  Optional.
    threshold
        Value in data units according to which the colors from textcolors are
        applied.  If None (the default) uses the middle of the colormap as
        separation.  Optional.
    **kwargs
        All other arguments are forwarded to each call to `text` used to create
        the text labels.
    """

    # Normalize the threshold to the images color range.
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.

    # Set default alignment to center, but allow it to be
    # overwritten by textkw.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)

    # Get the formatter in case a string is supplied
    if isinstance(valfmt, str):
        valfmt = matplotlib.ticker.StrMethodFormatter(valfmt)

    # Loop over the data and create a `Text` for each "pixel".
    # Change the text's color depending on the data.
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
#             kw.update(color=textcolors[int(im.norm(data[i, j]) > threshold)])
            text = im.axes.text(j+0.5, i+0.5, valfmt(data[i, j], None), **kw)#j+0.1, i+0.5
            texts.append(text)

    return texts
```

### filtered results

```python
# filtered results
rb_df = pd.DataFrame(data=np.zeros((len(celltypes), len(celltypes))), 
                     columns=celltypes, index=celltypes)
rbse_df = pd.DataFrame(data=np.zeros((len(celltypes), len(celltypes))), 
                     columns=celltypes, index=celltypes)
rbpvalue_df = pd.DataFrame(data=np.zeros((len(celltypes), len(celltypes))), 
                     columns=celltypes, index=celltypes)
numcoeqtl_df = pd.DataFrame(data=np.zeros((len(celltypes), len(celltypes))), 
                     columns=celltypes, index=celltypes)
num_anno_df = pd.DataFrame(data=np.zeros((len(celltypes), len(celltypes))), 
                     columns=celltypes, index=celltypes)
rbse_anno_df = pd.DataFrame(data=np.zeros((len(celltypes), len(celltypes))), 
                     columns=celltypes, index=celltypes)
for discovery_celltype in celltypes:
    # replication in other celltypes
    for replication_celltype in celltypes:
        if discovery_celltype != replication_celltype:
            rb_results = filtered_res_df[(filtered_res_df['celltype_discovery'] == discovery_celltype) &
                                         (filtered_res_df['celltype_replication'] == replication_celltype)]
            replicated_coeqtls_num = pd.read_csv(
                workdir/f'output/filtered_results/rb_calculations/discovery_{discovery_celltype}_replication_{replication_celltype}.tsv.gz',
                compression='gzip',
                sep='\t',
                index_col=0
            ).shape[0]
            if rb_results['r'].values[0] < 10 and discovery_celltype != 'B':
                rb_df.loc[replication_celltype, discovery_celltype] = rb_results['r'].values[0]
                rbse_df.loc[replication_celltype, discovery_celltype] = rb_results['se_r'].values[0]
                rbpvalue_df.loc[replication_celltype, discovery_celltype] = rb_results['p'].values[0]
                numcoeqtl_df.loc[replication_celltype, discovery_celltype] = replicated_coeqtls_num
                rbvalue = rb_results['r'].values[0]
                rbsevalue = rb_results['se_r'].values[0]
                num_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={replicated_coeqtls_num}"
                rbse_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"{rbvalue:.2f}\nN={replicated_coeqtls_num}"
            elif discovery_celltype == 'B':
                rb_df.loc[replication_celltype, discovery_celltype] = np.nan
                rbse_df.loc[replication_celltype, discovery_celltype] = np.nan
                rbpvalue_df.loc[replication_celltype, discovery_celltype] = 0
                numcoeqtl_df.loc[replication_celltype, discovery_celltype] = replicated_coeqtls_num
                rbvalue = rb_results['r'].values[0]
                rbsevalue = rb_results['se_r'].values[0]
                num_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={replicated_coeqtls_num}"
                rbse_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={replicated_coeqtls_num}"
            else:
                rb_df.loc[replication_celltype, discovery_celltype] = np.nan
                rbse_df.loc[replication_celltype, discovery_celltype] = np.nan
                rbpvalue_df.loc[replication_celltype, discovery_celltype] = 0
                numcoeqtl_df.loc[replication_celltype, discovery_celltype] = replicated_coeqtls_num
                num_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={replicated_coeqtls_num}"
                rbse_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={replicated_coeqtls_num}"
        else:
            rb_df.loc[replication_celltype, discovery_celltype] = 1
            rbse_df.loc[replication_celltype, discovery_celltype] = 0
            rbpvalue_df.loc[replication_celltype, discovery_celltype] = 0
            replicated_coeqtls_num = pd.read_csv(
                workdir/f'output/filtered_results/UT_{discovery_celltype}/coeqtls_fullresults_fixed.sig.tsv.gz',
                compression='gzip',
                sep='\t'
            ).shape[0]
            numcoeqtl_df.loc[replication_celltype, discovery_celltype] = replicated_coeqtls_num
            num_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={replicated_coeqtls_num}"
            rbse_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={replicated_coeqtls_num}"
    
replicated_ratio_df = pd.DataFrame(data=np.zeros((len(celltypes), len(celltypes))), 
                                   columns=celltypes, index=celltypes)
for discovery_celltype in numcoeqtl_df.columns:
    for replication_celltype in numcoeqtl_df.index:
        replicated_ratio_df.loc[replication_celltype, discovery_celltype] = \
        numcoeqtl_df.loc[replication_celltype, discovery_celltype] / numcoeqtl_df.loc[discovery_celltype, discovery_celltype]
```

### Sub celltypes in monocytes

```python
filtered_mono_res_df = pd.read_csv(workdir/'output/filtered_results/rb_calculations/monocyte_subcelltypes/summary.csv', 
                                   index_col=0)
filtered_mono_res_df
```

```python
# filtered results
mono_subcelltypes = ['monocyte', 'cMono', 'ncMono']
monorb_df = pd.DataFrame(data=np.zeros((len(mono_subcelltypes), len(mono_subcelltypes))), 
                     columns=mono_subcelltypes, index=mono_subcelltypes)
monorbse_df = pd.DataFrame(data=np.zeros((len(mono_subcelltypes), len(mono_subcelltypes))), 
                       columns=mono_subcelltypes, index=mono_subcelltypes)
monorbpvalue_df = pd.DataFrame(data=np.zeros((len(mono_subcelltypes), len(mono_subcelltypes))), 
                       columns=mono_subcelltypes, index=mono_subcelltypes)
mononumcoeqtl_df = pd.DataFrame(data=np.zeros((len(mono_subcelltypes), len(mono_subcelltypes))), 
                       columns=mono_subcelltypes, index=mono_subcelltypes)
monoanno_df = pd.DataFrame(data=np.zeros((len(mono_subcelltypes), len(mono_subcelltypes))), 
                       columns=mono_subcelltypes, index=mono_subcelltypes)
mononum_anno_df = pd.DataFrame(data=np.zeros((len(mono_subcelltypes), len(mono_subcelltypes))), 
                     columns=mono_subcelltypes, index=mono_subcelltypes)

for discovery_celltype in mono_subcelltypes:
    # replication in other celltypes
    for replication_celltype in mono_subcelltypes:
        if discovery_celltype != replication_celltype:
            monorb_results = filtered_mono_res_df[(filtered_mono_res_df['celltype_discovery'] == discovery_celltype) &
                                         (filtered_mono_res_df['celltype_replication'] == replication_celltype)]
            monoreplicated_coeqtls_num = pd.read_csv(
                workdir/f'output/filtered_results/rb_calculations/monocyte_subcelltypes/discovery_{discovery_celltype}_replication_{replication_celltype}.tsv.gz',
                compression='gzip',
                sep='\t',
                index_col=0
            ).shape[0]
            if monorb_results['r'].values[0] < 10:
                monorb_df.loc[replication_celltype, discovery_celltype] = monorb_results['r'].values[0]
                monorbse_df.loc[replication_celltype, discovery_celltype] = monorb_results['se_r'].values[0]
                monorbpvalue_df.loc[replication_celltype, discovery_celltype] = monorb_results['p'].values[0]
                mononumcoeqtl_df.loc[replication_celltype, discovery_celltype] = monoreplicated_coeqtls_num
                monorbvalue = monorb_results['r'].values[0]
                monorbsevalue = monorb_results['se_r'].values[0]
                monoanno_df.loc[replication_celltype, discovery_celltype] = \
                f"rb={monorbvalue:.2f}\nN={monoreplicated_coeqtls_num}"
                mononum_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={monoreplicated_coeqtls_num}"
            else:
                monorb_df.loc[replication_celltype, discovery_celltype] = np.nan
                monorbse_df.loc[replication_celltype, discovery_celltype] = np.nan
                monorbpvalue_df.loc[replication_celltype, discovery_celltype] = 0
                mononumcoeqtl_df.loc[replication_celltype, discovery_celltype] = monoreplicated_coeqtls_num
                monoanno_df.loc[replication_celltype, discovery_celltype] = \
                f"rb=NA\nN={monoreplicated_coeqtls_num}"
                mononum_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={monoreplicated_coeqtls_num}"
        else:
            monorb_df.loc[replication_celltype, discovery_celltype] = 1
            monorbse_df.loc[replication_celltype, discovery_celltype] = 0
            monorbpvalue_df.loc[replication_celltype, discovery_celltype] = 0
            monoreplicated_coeqtls_num = pd.read_csv(
                workdir/f'output/filtered_results/UT_{discovery_celltype}/coeqtls_fullresults_fixed.sig.tsv.gz',
                compression='gzip',
                sep='\t'
            ).shape[0]
            mononumcoeqtl_df.loc[replication_celltype, discovery_celltype] = monoreplicated_coeqtls_num
            monoanno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={monoreplicated_coeqtls_num}"
            mononum_anno_df.loc[replication_celltype, discovery_celltype] = \
                f"N={monoreplicated_coeqtls_num}"
    
monoreplicated_ratio_df = pd.DataFrame(data=np.zeros((len(mono_subcelltypes), len(mono_subcelltypes))), 
                       columns=mono_subcelltypes, index=mono_subcelltypes)
for discovery_celltype in mononumcoeqtl_df.columns:
    for replication_celltype in mononumcoeqtl_df.index:
        monoreplicated_ratio_df.loc[replication_celltype, discovery_celltype] = \
        mononumcoeqtl_df.loc[replication_celltype, discovery_celltype] / mononumcoeqtl_df.loc[discovery_celltype, discovery_celltype]
```
